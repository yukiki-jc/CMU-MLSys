```python
import gzip
import struct
import numpy as np
```

```python
image_filename = "../data/train-images-idx3-ubyte.gz"
images = []
label_filename = "../data/train-labels-idx1-ubyte.gz"
def read_header(f):
    int_bytes = 4
    # read image
    magic_num_buf = f.read(int_bytes)
    magic_num = struct.unpack('>i', magic_num_buf)[0]
    dim = magic_num & 0x000000ff
    int_buf = f.read(int_bytes * dim)
    pack_str = '>' + 'i' * dim
    dims = struct.unpack(pack_str, int_buf)
    return dims

with gzip.open(image_filename, 'rb') as f:
    dims = read_header(f)
    print(dims)
    for i in range(dims[0]):
        length = dims[1] * dims[2]
        buf = f.read(length)
        pack_str = 'B' * length
        r = list(struct.unpack(pack_str, buf)) 
        images.append(r)

with gzip.open(label_filename, 'rb') as f:
    dims = read_header(f)
    print(dims)
    length = dims[0]
    buf = f.read(length)
    pack_str = 'B' * length
    r = list(struct.unpack(pack_str, buf))
images[0][10] = 253
print(np.array(images[0], dtype=np.float32) / 255)
print(len(r))
```
